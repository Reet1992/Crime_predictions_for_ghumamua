```python
#### decomposition 



#### Decomposition #### 

import warnings
import itertools
warnings.filterwarnings("ignore")
import statsmodels.api as sm

from pylab import rcParams
rcParams['figure.figsize'] = 18, 8
```

```python
df2.index
```

```python
df3 = df2['Total Incidents'].resample('MS').mean()
```

```python
df3.head()
```

```python
df3 = pd.DataFrame(df3)
```

```python
df3.head()
```

```python
df3.plot(figsize=(15, 6))
plt.show()
```

```python
from statsmodels.tsa import seasonal

decomposed1 = seasonal.seasonal_decompose(x=df3, freq=22, extrapolate_trend='freq',model = 'additive')
fig = decomposed1.plot()
plt.show()
```

```python
#### Time Series prediction with ARIMA 


p = d = q = range(0, 2)
pd = list(itertools.product(p, d, q))
seasonal_pd = [(x[0], x[1], x[2],16) for x in list(itertools.product(p, d,q))]

print(seasonal_pd)

for param in pd:
    for param_seasonal in seasonal_pd:        
        mod = sm.tsa.statespace.SARIMAX(df3['Total Incidents'])
results = mod.fit()
print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))

### Fitting with the latest Value

mod = sm.tsa.statespace.SARIMAX(df3['Total Incidents'])
results = mod.fit()

print(results.summary())


#### plot_result_diagnostic statistics ##

results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
#### Getting the BEst paramters #### 



for param in pd:
    for param_seasonal in seasonal_pd:
            mod = sm.tsa.statespace.SARIMAX(df3['Total Incidents'],order=param,seasonal_order=(0,0,0,0))
            results = mod.fit()
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
```

```python
mod = sm.tsa.statespace.SARIMAX(df3['Total Incidents'],order=(1, 1, 0),seasonal_order=(0,0,0,16))
results = mod.fit()

print(results.summary())

results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
len(df3)
```

```python
pred = results.get_prediction(start=df3.index[50], dynamic=False)
pred_ci = pred.conf_int()
```

```python
pred_ci.iloc[:,1]
```

```python
ax = df3['2016-12-01':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))

ax.fill_between(pred_ci.index,pred_ci.iloc[:, 1],pred_ci.iloc[:, 0], color='k', alpha=.2)


ax.set_xlabel('Date')
ax.set_ylabel('No_of_Total_incidents')
plt.legend()
plt.show()
```
